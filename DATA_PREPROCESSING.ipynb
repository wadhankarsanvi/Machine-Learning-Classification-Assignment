```python
import pandas as pd
df = pd.read_csv('obesity.csv')
```

```python
df.duplicated()
```

df.drop_duplicates(subset = None, keep = first, inplace = False,ignore_index = False)


subset(deafult = NOne):


specifies which column to consider for identifying duplicates
if None, all columns are used
if a list of column names is provided, duplicates are deteced based on those columns.

keep(default = 'first'):

determines which duplicate row to keep.
'first':  Keeps the first ocurence and drops the rest.
'last': keeps the last occurence an drops others

false removes all the duplicate rows.

(inplace = false): ko rakhna padega ya fir pura hi duplicate remov kardega


if true it modifies the dataframe in place and returns None
if false, returns a new dataframe with duplicates removed


ignore_index(default = False):


if True, resets the index in the resulting Data frame

```python
df.drop_duplicates(subset = None, keep = 'first', inplace = False,ignore_index = False)
# #false removes all the duplicate rows, inplace = false ko rakhna padega ya fir pura hi duplicate remov kardega
# if true it modifies the dataframe in place and returns None
# ignore_index(default = False)
```

DETECTION AND REMOVAL OF OUTLIER


```python
import sklearn
from sklearn.datasets import load_diabetes
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diabetes = load_diabetes()
```

```python
column_name = diabetes.feature_names
df_diabetes = pd.DataFrame(diabetes.data,columns = column_name)
```

```python
sns.boxplot(df_diabetes['bmi'])
plt.title('Boxplot of BMI')
plt.show()
```

```python
def removal_box(df, column,threshold):
  remove_outlier = df[df[column]<=threshold] # we set a threshold value accordint to the dataset and then remove the values which are greater than the threshold value
  sns.boxplot(remove_outlier[column])
  plt.title(f'Boxplot without Outlier of {column}')
  plt.show()
  return remove_outlier
threshold = 0.12

no_outliers = removal_box(df_diabetes,'bmi',threshold)
```

```python
#Zscore => X1 = (X - mean )/ (Standard deviation)
#scaling of age to get a definite range of all attributes so that we can work on all of them in the provided range
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(df_diabetes['age']))
print(z)
```
